```python
#importing data from csv file
import pandas as pd
df = pd.read_csv("NABIL.csv")
df
```

```python
df = df[['Date', 'Close']]
df
```

```python
df['Date']
```

```python
#Since the datatype of date is object, we have to convert it into dateTime
import datetime
def str_to_datetime(s):
    split = s.split('/')
    month, day, year = int(split[0]), int(split[1]), int(split[2])
    return datetime.datetime(year = year, month = month, day = day)

# datetime_object = str_to_datetime('12/2/2024')
# datetime_object
```

```python
# Applying the str_to_datetime functon in all the dates of that database
df['Date'] = df['Date'].apply(str_to_datetime)
df['Date']
```

```python
#Removing the index of 0_to_1126, and set the index as date
df.index = df.pop('Date')
```
